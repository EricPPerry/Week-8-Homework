## Exploratory Data Analysis

> Use Matplotlib to display your data analysis

### Which production companies released the most movies in the last 10 years? Display the top 5 production companies.

```python
df.groupby('production_companies').count().sort_values('id',ascending=False)
top_5_production = df.groupby('production_companies', as_index=False).count().sort_values('id',ascending=False)[['production_companies','id']].head()
top_5 = []
for company in range(len(top_5_production)):
    top_5.append(top_5_production.iloc[company]['production_companies'])

top_5_count = []
for company in range(len(top_5_production)):
    top_5_count.append(top_5_production.iloc[company]['id'])

plt.figure(figsize=(12,6))
plt.ylabel('Movies Released in Last 10 Years')
plt.xlabel('Production Company')
plt.bar(top_5,top_5_count)
```
